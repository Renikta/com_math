```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
alpha  = 2.0
beta   = 0.0015
gamma  = 5.0
sigma0 = 3.0
phi0   = 0.0525
C      = 5.0
Ka1    = 0.05
Ka2    = 0.35
```

```python
def f1(sigma, phi):
    return (alpha * sigma * sigma) / (sigma + sigma0) - Ka1 * sigma - gamma * sigma * phi

def f2(sigma, phi):
    return beta * sigma * (1 - phi / C) * (1 + (phi/phi0)**2) - Ka2 * phi
```

```python
def norma(sigma, phi):
    return math.sqrt(sigma ** 2 + phi ** 2)

def step_one(sigma, phi, f1, f2, h):
    sigma_next_iter = sigma + h * f1(sigma, phi)
    phi_next_iter = phi + h * f2(sigma, phi)
    return sigma_next_iter, phi_next_iter

def step_two(sigma, phi,sigma_next_iter, phi_next_iter, f1, f2, h):
    sigma_next = sigma + h * (f1(sigma, phi) + f1(sigma_next_iter, phi_next_iter)) / 2
    phi_next = phi + h * (f2(sigma, phi) + f2(sigma_next_iter, phi_next_iter)) / 2
    return sigma_next, phi_next


def one_iteration(sigma, phi, f1, f2, h, accuracy):
    sigma_next_iter, phi_next_iter = step_one(sigma, phi, f1, f2, h)
    
    sigma_next, phi_next = step_two(sigma, phi,sigma_next_iter, phi_next_iter, f1, f2, h)

    while norma(sigma_next - sigma_next_iter, phi_next - phi_next_iter) > accuracy:
        sigma_next_iter = sigma_next
        phi_next_iter = phi_next
        sigma_next, phi_next = step_two(sigma, phi,sigma_next_iter, phi_next_iter, f1, f2, h)
    
        
    return sigma_next, phi_next
```

```python
h = 0.0001
interval = 20
size = int(interval / h + 1)

time = [0] * size

for i in range(size):
    time[i] = h * i

sigma = [0] * size
sigma[0] = sigma0

phi = [0] * size
phi[0] = phi0
```

```python
for i in range(size - 1):
    sigma[i + 1], phi[i + 1] = one_iteration(sigma[i], phi[i], f1, f2, h, 0.001)


plt.figure(figsize=[12, 4])
plt.plot(time, sigma, label = 'sigma = f(t)')

plt.title("sigma = f(t)")
plt.xlabel("$t$")
plt.ylabel("$y1$")
plt.grid()
plt.legend(loc = 'best', bbox_to_anchor = (1, 1))
plt.tight_layout() 
plt.show()


plt.figure(figsize=[12, 4])
plt.plot(time, phi, label = 'y2 = f(t)')

plt.title("y2 = f(t)")
plt.xlabel("$t$")
plt.ylabel("$y2$")
plt.grid()
plt.legend(loc = 'best', bbox_to_anchor = (1, 1))
plt.tight_layout() 
plt.show()


plt.figure(figsize = [12, 4])
plt.plot(sigma, phi, label = 'y2 = f(y1)')
```

```python
# Насилие над параметрами
# Проварьируем alpha

maximum = gamma * C

step = 1
size_a = int(maximum / step)

#plt.figure(figsize = [12, 4])

for i in range(size_a):
    alpha = step * i

    for i in range(size - 1):
        sigma[i + 1], phi[i + 1] = one_iteration(sigma[i], phi[i], f1, f2, h, 0.001)

    plt.plot(sigma, phi)
```
