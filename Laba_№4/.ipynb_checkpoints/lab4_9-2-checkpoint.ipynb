```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f1(t, y1, y2):
    return - p * t * math.cos(y2)

def f2(t, y1, y2):
    return y1
```

```python
# Используем неявный метод Рунге-Кутты

def norm(x, y):
    return math.sqrt(x ** 2 + y ** 2)

def expectation(t, y1, y2, f1, f2, h):
    
    y1_n_iter = y1 + h * f1(t, y1, y2)
    y2_n_iter = y2 + h * f2(t, y1, y2)
    
    return y1_n_iter, y2_n_iter


def correction(t, y1, y2, t_n, y1_n_iter, y2_n_iter, f1, f2, h):
    
    y1_n = y1 + h * (f1(t, y1, y2) + f1(t_n, y1_n_iter, y2_n_iter)) / 2
    y2_n = y2 + h * (f2(t, y1, y2) + f2(t_n, y1_n_iter, y2_n_iter)) / 2
    
    return y1_n, y2_n


def find_next_step(t, y1, y2, t_n, f1, f2, h, accuracy):
    
    y1_n_iter, y2_n_iter = expectation(t, y1, y2, f1, f2, h)
    y1_n, y2_n = correction(t, y1, y2, t_n, y1_n_iter, y2_n_iter, f1, f2, h)

    while norm(y1_n - y1_n_iter, y2_n - y2_n_iter) > accuracy:
        y1_n_iter = y1_n
        y2_n_iter = y2_n
        y1_n, y2_n = correction(t, y1, y2, t_n, y1_n_iter, y2_n_iter, f1, f2, h)
    
    return y1_n, y2_n
```

```python
def solve_by_shooting_method(x, z, y, z0, y_des_right_limit, accuracy):
    z[0] = z0  #
    
    # Пусть A0 = 0
    y_left_limit_1 = 0
    y_left_limit_2 = 0
    
    y[0] = y_left_limit_1
    
    # Ищем первое решение
    
    for i in range(size - 1):
        z[i + 1], y[i + 1] = find_next_step(x[i], z[i], y[i], x[i + 1], f1, f2, h, 0.001)
        
    y_right_limit_1 = y[-1] # Зафиксировали полученное значение справа
    
    # Ищем значение A1 по смене знака
    y_right_limit_2 = y_right_limit_1  # Берем в качестве начального значение right1
    j = 1
    
    
    #############(            не меняется            )###########(              меняется             )
    while np.sign(y_right_limit_1 - y_des_right_limit) == np.sign(y_right_limit_2 - y_des_right_limit):
        y_left_limit_2 = j
        y[0] = y_left_limit_2  # Начальное значение
        
        for i in range(size - 1):
            z[i + 1], y[i + 1] = find_next_step(x[i], z[i], y[i], x[i + 1], f1, f2, h, 0.001)
        
        y_right_limit_2 = y[-1]  # Записываем результат
        j += 1
    
    
    # Начальное значение для ввода в цикл
    new_y_right_limit = y_des_right_limit + accuracy * 2
    
    # Уточняем решение методом бинарного поиска
    while abs(new_y_right_limit - y_des_right_limit) > accuracy:
        
        # Уменьшаем границу
        new_y_left_limit = (y_left_limit_1 + y_left_limit_2) / 2
        y[0] = new_y_left_limit
        
        for i in range(size - 1):
            z[i + 1], y[i + 1] = find_next_step(x[i], z[i], y[i], x[i + 1], f1, f2, h, 0.001)
        
        new_y_right_limit = y[-1] # Фиксируем новую границу по правую сторону
        
        if np.sign(new_y_right_limit) != np.sign(y_right_limit_1): # По знаку определяем, где лежит нужное значение
            y_left_limit_2 = new_y_left_limit
        else:
            y_left_limit_1 = new_y_left_limit
```

```python
p_arr = [1, 4, 7, 25, 50, 100, 200, 500] # more parameter values are used in order to get different pictures on the graph
max_x = 1    # 1 second
accuracy = 0.1
h = 0.01
size = int(max_x / h + 1)
z0 = 0

x = [0] * size
for i in range(size):
    x[i] = h * i

z = [0] * size
z[0] = 0

y = [0] * size
y[0] = 0
```

```python
for p in p_arr:
    print("p = ", p)

    solve_by_shooting_method(x, z, y, z0, 0, accuracy)  # y(1) = 0, accuracy = 0.001
    
    # y = f(x)
    plt.figure(figsize = [12, 4])
    plt.plot(x, y, label = 'y = f(x)')
    plt.title("y = f(x)")
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    plt.grid()
    plt.legend(loc = 'best', bbox_to_anchor=(1, 1))
    plt.tight_layout() 
    plt.show()
    print("y[1] = ", y[-1])
```
