```python
# Решение явным методом Рунге-Кутты 4 порядка

interval = 25 
h        = 0.005
size = int(interval / h + 1)

k_1 = [0] * 2
k_2 = [0] * 2
k_3 = [0] * 2
k_4 = [0] * 2

u = [0] * size
v = [0] * size

u[0] = x0
v[0] = y0

u_ = [0] * size
v_ = [0] * size


def one_iteration_4(u, v, B):
    
    k_1[0] = h * f1(u, v, B)
    k_1[1] = h * f2(u, v, B)
    
    k_2[0] = h * f1(u + h/2, v + k_1[0]/2, B)
    k_2[1] = h * f2(u + h/2, v + k_1[1]/2, B)
    
    k_3[0] = h * f1(u + h/2, v + k_2[0]/2, B)
    k_3[1] = h * f2(u + h/2, v + k_2[1]/2, B)
    
    k_4[0] = h * f1(u + h, v + k_3[0], B)
    k_4[0] = h * f2(u + h, v + k_3[0], B)
    
    
    vn = v + (k_1[0] + 2 * k_2[0] + 2 * k_3[0] + k_4[1]) / 6
    un = u + (k_1[1] + 2 * k_2[1] + 2 * k_3[1] + k_4[1]) / 6
    
    return un, vn
```

```python
for B in range(B_min, B_max + 1, 1):

    for i in range(size - 1):
        u[i + 1], v[i + 1] = one_iteration_4(u[i], v[i], B);
    
    for i in range(size):
        u_[i] = f1(u[i], v[i], B)
        v_[i] = f2(u[i], v[i], B)

    time = [0] * size
    for i in range(size):
        time[i] = h * i
        
    print("B = ", B)

    plt.figure(figsize = [8, 6])
    plt.plot(time, u, label = 'u(t)')
    plt.plot(time, v, label = 'v(t)')
    plt.xlabel("TIME t")
    plt.ylabel("")
    plt.title("Метод Рунге-Кутта 4 порядка")
    plt.tight_layout() 
    plt.grid()
    plt.legend(loc = 'best', bbox_to_anchor = (1, 1))
    plt.show()
    
    plt.figure(figsize = [8, 6])
    plt.plot(v, v_, label = 'v(t)')
    plt.xlabel("v")
    plt.ylabel("v'")
    plt.title("Фазовый портрет. Метод Рунге-Кутта 1 порядка.")
    plt.tight_layout() 
    plt.grid()
    plt.legend(loc = 'best', bbox_to_anchor = (1, 1))
    plt.show()
```
